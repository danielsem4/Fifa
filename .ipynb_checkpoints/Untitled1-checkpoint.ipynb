```python
import numpy as np
import pandas as pd
```

```python
temp_df = pd.read_csv("Players.csv")
```

### data cleaning part 🧹

```python
df = temp_df.drop(['Unnamed: 0', 'Name','Version', 'Price', 'WR', 'Game_Stats', 'Base_Stats', 'Popularity'],axis=1) 
df = df.drop_duplicates()
# Getting rid of unnecessary couloms and dropping duplicates
```

```python
split_height = df["Height"].str.split("cm",n=1,expand=True)
split_height.rename(columns={0:'Player_Height',1:'Delete'})
# I created a new table to separate the height from the rest of the string
```

```python
df[['Player_Height','Delete']] = df['Height'].str.split("cm",n=1,expand=True)
df = df.drop(['Delete', 'Height'],axis=1)
df['Player_Height'] = df['Player_Height'].str.replace(r'\D', '').astype(int)
df['Player_Height'].describe(include='all')
# Added the new table to the main table and deleted the unnecessary columns
```

```python
df['Weight'] = df['Weight'].str.replace(r'\D', '').astype(int)
# The string was removed from the weight
```

```python
df.dropna(inplace=True)
df['Aggression'] = df['Aggression'].astype(int)
df['Balance'] = df['Balance'].astype(int)
df['Ball_Control'] = df['Ball_Control'].astype(int)
df['Jumping'] = df['Jumping'].astype(int)
df['Long_Passing'] = df['Long_Passing'].astype(int)
df['Stamina'] = df['Stamina'].astype(int)
df['Player_Height'] = df['Player_Height'].astype(int)
# Changed the type to int
```

```python
df.drop(df.loc[df["Position"] == 'GK'].index, inplace=True)
# Removed the goalkeepers from the table
```

```python
df.describe(include='all')
```

```python
df.dropna(inplace=True)
```

```python
df.info()
```

```python
position_dict = {"ST":1 ,"CF":1 ,"WR":2, "RW":2, "LF":2, "LW":2, "CAM":3, "CM":4, "LM":4,
                 "RM":4, "CDM":5, "LB":6, "RB":6, "LWB":6, "RWB":6, "CB":7}
df['Position'] = df['Position'].replace(position_dict)
```

# outliers handaling ➕ ➖ ➗ ✖️


```python
df2=df.copy()
mycols=df2.select_dtypes('number').columns
for col in mycols:        
    Q1 = np.percentile(df2[col], 25)
    Q3 = np.percentile(df2[col], 75)
    IQR = Q3 - Q1
    IQR_range = 1.5 * IQR
    df2.loc[(df2[col] < Q1 - IQR_range) | (df2[col] > Q3 + IQR_range ),col]=np.nan
```

```python
df2.info()
```

```python
df2.dropna()
```

```python
df3=df.copy()
mycols=df3.select_dtypes('number').columns
for col in mycols:
    z_score = (df3[col] - df3[col].mean()) / df3[col].std()
    print(df3[col] - df3[col].mean())
    outliers = abs(z_score) > 3
    df3.loc[outliers,col]=np.nan
```

```python
df3.info()
```

```python
df3.dropna()
```

```python
df.to_csv("Renewed_Players.csv")
df3.to_csv("Players_Without_Outliers.csv")
```
